### Обработка пропусков столбцов `balance` и `age`

```python
df['is_null'] = df.isnull().any(axis=1).astype(int)
df.head()
```

```python
# Создаем сводную таблицу, которая покажет средние значения
# зарплаты и возраста по пропущенным и заполенным строкам
pivot_table = df.pivot_table(values= ['est_salary', 'age'],
                             index='is_null',
                             aggfunc=pd.Series.mean)
# Вычисляем относительную разницу между первой и второй строкой каждого столбца
pct_change = pivot_table.pct_change(periods=1).iloc[1:2] * 100

# Переименовываем строки
pct_change = pct_change.rename(index={pct_change.index[0]: 'Относительная разница, в %'})
pivot_table.rename(index={1: 'значение пропущено'}, inplace=True)
pivot_table.rename(index={0: 'значение не пропущено'}, inplace=True)
# Добавляем относительную разницу к сводной таблице
pivot_table = pivot_table.append(pct_change)


# Создаем сводную таблицу, которая покажет медианные значения
# количества собственности и продуктов по пропущенным и заполенным строкам
pivot_table_one = df.pivot_table(values= ['equity', 'products'],
                             index='is_null',
                             aggfunc=pd.Series.median)
# Вычисляем относительную разницу между первой и второй строкой каждого столбца
pct_change_one = pivot_table_one.pct_change(periods=1).iloc[1:2] * 100
# Переименовываем строки
pct_change_one = pct_change_one.rename(index={pct_change_one.index[0]: 'Относительная разница, в %'})
pivot_table_one.rename(index={1: 'значение пропущено'}, inplace=True)
pivot_table_one.rename(index={0: 'значение не пропущено'}, inplace=True)
# Добавляем полученные значения к сводной таблице
pivot_table_one = pivot_table_one.append(pct_change_one)

# Создаем сводную таблицу, которая покажет моды
# города и пола по пропущенным и заполенным строкам
pivot_table_two = df.pivot_table(values= ['city', 'gender'],
                             index='is_null',
                             aggfunc=pd.Series.mode)
# Переименовываем строки
pivot_table_two.rename(index={1: 'значение пропущено'}, inplace=True)
pivot_table_two.rename(index={0: 'значение не пропущено'}, inplace=True)

# Объединяем все сводные таблицы при помощи concat
# по столбцам
all_pivot = pd.concat([pivot_table, pivot_table_one, pivot_table_two], axis=1)
# Переименовываем столбцы
all_pivot.rename(columns = {'age':'средний возраст', 'est_salary':'средняя зарплата',
                           'equity':'медианное количество собственности',
                           'products':'медианное количество продуктов',
                           'city':'мода по городу', 'gender':'мода по полу'}, inplace = True)
# Сранительная таблица готова
all_pivot.T
```

```python
# Пропуски в столбцах age и balance одновременно
df[df['balance'].isnull() & df['age'].isnull()]
```

Пропуски в столбцах age и balance одновременно следует удалить, потому что в таких строках отсутствуют два признака и от этого выборка может быть неменее достоверной. Таких случаев всего 16, поэтому действительная картина не исказится.

```python
# Исключаем 16 пропущенных значений одновременно в balance и age
# С помощью знака ~ исключаем из датасета строки
# С помощью знака & делаем так, чтоб условия удовлетворялись одновременно
df = df[~(df['balance'].isnull() & df['age'].isnull())]
```

```python
# Оцениваем количество пропусков в balance
print('Количество пропусков в столбце balance: ', df['balance'].isnull().sum())
```

Количество пропущенных значений в столбце `balance` составляет примерно 23 %. Эти пропуски можно классифицировать как `MNAR (Missing Not At Random)`, так как переменные, которые находятся в датасете не объясняют причину пропусков. Точные причины и природу этих пропусков опознать невозможно, от чего самым разумным решением будет либо проигнорировать пропуски, либо заполнить числовой заглушкой, например -1, но при этом фильтровать ее, что бы избежать выбросов и искажения значений корреляций. Удалять пропущенные значения не имеет смысла, так как мы лишимся огромного пласта данных. Заменять пропуски на 0 - тоже плохой вариант, так как это испортит статистическую выборку.

```python
# Удаление пропусков в столбце balance
df.dropna(subset=['balance'])
```

```python
# Оцениваем количество пропусков в age
print('Количество пропусков в столбце age: ', df['age'].isnull().sum())
```

Аналогичная ситуация с пропусками в столбце `age`. Хоть их и примерно в 100 раз меньше, чем пропусков в столбце `balance`, то имеет смысл поставить заглушку -1 и в дальнейшем фильтровать.

```python
# Удаление пропусков в столбце age
df.dropna(subset=['age'])
```

## Статистический анализ данных

### Формулировка и проверка гипотезы № 1

Так как оценочная зарплата является одним из признаком оттока клиентов, то на ее основе сформулируем нулевую и альтернативную гипотезы:

1. H₀: все отточные и неотточные клиенты имеют одинаковую оценочную зарплату.
2. H₁: все отточные и неотточные клиенты имеют разную оценочную зарплату.

Проверка гипотез будет осуществляться либо с помощью `t-test'a`, либо с помощью `u-test'a`, так как мы будем сравнивать средние две независимые руппы: отточных и неотточных клиентов. Выбор оптимального теста будет зависеть от нормальности распределения данных в столбце `est_salary`. Если данные распределены нормально, то будет использоваться `t-test`, если ненормально, то `u-test` Манна-Уитни, так как он более устойчив к выбросам и отклонениям от нормальности данных, что делает его предпочтительным выбором в таких случаях.

Для проверки нормальности распределения будет использоваться тест Шапиро-Уилка:

```python
# Тест Шапиро-Уилка для оценки нормальности распределения
# в столбце est-salary
stat, p = stats.shapiro(df['est_salary'])
print('Тест Шапиро-Уилка: статистика = {}, p-value = {}'.format(stat, p))
```

Данные теста показали, что значение p-value меньше значения alpha = 0.05. Это означает, что распределение в столбце `est_salary` является ненормальным, а значит более разумным вариантом будет использоваться `u-test` Манна-Уитни для проверки гипотезы:

```python
# Создаем два списка данных для отточных и неотточных клиентов
churn_0 = df[df['churn']==0]['est_salary']
churn_1 = df[df['churn']==1]['est_salary']

# Сохраняем результаты теста в переменной
results = stats.mannwhitneyu(churn_0, churn_1, alternative='two-sided')

# Задаем alpha
alpha = 0.05

# Смотрим дисперсии. Если отличаются, то equal_var = False:
variance_0 = np.var(churn_0)
variance_1 = np.var(churn_1)

print("Дисперсия 0:", variance_0)
print("Дисперсия 1:", variance_1)

# вывод значения средних значений на экран
print('Среднее значение не отточных клиентов: ', churn_0.mean())
print('Среднее значение отточных клиентов: ', churn_1.mean())

# вывод значения p-value на экран 
print ('P-value: ', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print ('Отвергаем нулевую гипотезу')
else:
    print ('Не получилось отвергнуть нулевую гипотезу')
```

### Формулировка и проверка гипотезы № 2

Сформулируем нулевую и альтернативную гипотезы:

1. H₀: все отточные клиенты одного пола.
2. H₁: все отточные клиенты имеют разный пол.

Проверка гипотез будет осуществляться с помощью `z-test'a (тест пропорций)`, так как мы будем сравнивать средние две зависимые (парные) группы: отточных клиентов, которые оличаются по половому признаку.

```python
# Создаем два списка данных для отточных и неотточных клиентов
male = df[(df['churn']== 1) & (df['male']== 1)]['male'].count()
female = df[(df['churn']== 1) & (df['female']== 1)]['female'].count()

male_all = df.query('male == 1')['male'].count()
female_all = df.query('female == 1')['female'].count()
```

```python
male
```

```python
male_all
```

```python
female
```

```python
def z_test(successes1, successes2, trials1, trials2, alpha=0.05):
    p1 = successes1 / trials1
    p2 = successes2 / trials2
    
    print(successes1, successes2, trials1, trials2)

    p_combined = (successes1 + successes2) / (trials1 + trials2)
    difference = p1 - p2

    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1/trials1 + 1/trials2)) 
    distr = st.norm(0, 1)  
    p_value = (1 - distr.cdf(abs(z_value))) * 2 
    print('p-значение: ', p_value)

    if (p_value < alpha): 
        display('Отвергаем нулевую гипотезу, между выборками есть статистически значимые различия')
    else:
        display('Не получилось отвергнуть нулевую гипотезу, статистически значимых различий в выборках нет')
```

```python
z_test(male, female, male_all, female_all, alpha=0.05)
```

Согласно техническому заданию, необходимо сформировать сегменты, куда будут входить не менее 500 клиентов. Поэтому имеет смысл разделить всех отточных клиентов на `три сегмента`.

```python
def make_cluster(x, max_clasters):
    # Стандартизируем данные
    x_st = StandardScaler().fit_transform(x)

    # Построение дендрограммы
    linked = linkage(x_st, method='ward')

    plt.figure(figsize=(14,4))  
    dendrogram(linked, orientation='top', no_labels=True)
    plt.title('Дендограмма отточных клиентов банка', size=20, pad=30)
    plt.show()
    
    # создание модели K-Mens
    
    for i in range(3, max_clasters+1):

        # выводим статистику по средним значениям наших признаков по кластеру
        df['km_cluster'] = KMeans(n_clusters=i, random_state=0).fit_predict(x_st)


        # таблица средних значений, сгруппированная по кластерам
        display(df.groupby('km_cluster').mean())
```
