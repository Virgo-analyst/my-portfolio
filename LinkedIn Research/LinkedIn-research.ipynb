## Введение

LinkedIn является одним из ведущих онлайн-сообществ, объединяющим профессионалов со всего мира. Портал предоставляет уникальную возможность исследовать требования, навыки, и опыт, необходимые для успешной карьеры в области анализа данных. В рамках исследования будет проведен анализ различных аспектов вакансий для Junior Data Analyst на LinkedIn. Будут рассмотрены популярные сферы, в которых востребованы специалисты по анализу данных. Особое внимание уделено требованиям работодателей к кандидатам (hard skills), необходимые для успешной работы в данной сфере. Кроме того, будет осуществлен анализ типов занятости и локаций работы для Junior Data Analyst в странах и городах Европы. Результаты исследования, оформленные в виде дашборда, позволят профессиональным аналитикам, студентам и выпускникам сферы анализа данных принять более информированное решение о своей карьере и выборе работы за рубежом. Это исследование принесет практическую пользу и поможет определиться с дальнейшими шагами в карьерном развитии в сфере анализа данных.

## Цель исследования

Визуализировать информацию о рынке вакансий для Junior Data Analyst в Европе.

## Задачи исследования

Подготовка данных:
1. выгрузить датасет из LinkedIn и провести его обзор;
2. провести обработку данных: заполнить пропуски, проверить данные на дублирование, изменить тип данных при необходимости;
3. перевести данные на один язык (английский);
4. исключить из перечня вакансий тимлидов, сеньоров и миддлов;
5. вытащить из описания вакансии hard skills, которые работодатель требует от соискателя;
6. удалить лишние столбцы датафрейма, которые не пригодятся для визуализации информации.

Визуализация данных:
Построить интерактивный дашборд в любой из BI систем (с учетом возможности расшарить этот дашборд). По требованию от заказчика дашборд должен содержать:
1. `фильтры` — по стране и по типу занятости;

## Работа с пропусками

```python
# Знакомство со значениями столбца workplace:
df['workplace'].value_counts()
```

```python
# Замена пропусков на значение no datas:
df['workplace'] = df['workplace'].fillna('No datas')
df['workplace'].value_counts()
```

Пропущенные значения столбца `workplace` мы классифицируем как `MNAR`. Так как нам неизвестна природа этих пропусков, то самым оптимальным решением будет поставить заглушку в виде `No datas`.

```python
# Удаление пропущенных строк в столбцах company_name,
# industry, company_size
df = df.dropna(subset=['company_name', 'industry','company_size'])
```

Данный шаг был выполнен, поскольку столбцы нужны для дальнейших подсчетов. Если подобные вакансии с пропущенными значениями оставить, то они все равно не будут учитываться в визуализации, а само исследование выйдет некорректным, если, допустим, в одной строке столбец с названием компании будет заполнен, а с индустрией - нет.

Следующим шагом будет проанализировать столбец `applicants`. Из условия мы знаем, что здесь сохранено количество откликов на вакансии. Разумно предположить гипотезу о том, что пропуски связаны с тем, что на вакансию не откликались. Но прежде чем заполнить пропуски значением 0, убедимся, что такого значения нет в данных:

```python
df.query('applicants == 0')
```

```python
# Замена пропущенных значений в столбце applicants на 0
df['applicants'] = df['applicants'].fillna(0)
```

В столбце aplicants сохранено количество откликов на вакансию. Пропущенные значения в строках говорят о том, что на эту вакансию соискатели не откликались, поэтому будет разумным заменить их на 0.

<font color='#31708f'>📝 <b>Комментарий ревьюера:</b> По гипотезе возникновения пропусков все верно, но перед заменой на одно из значений, хорошо бы убедиться, что таких значений не было в данных. Т.е. в нашем случае проверить, есть ли 0 в этом столбце.

## Перевод текстовых данных на английский язык

Для перевода данных к единому языку воспользуемся средствами библиотеки `googletrans`:

```python
# Смотрим список вакансий
df['job_title'].value_counts()
```

Видно, что данные в датасете мультиязычные (по мимо английского - чешский, польский, немецкий, итальянский и т.д.), от того есть потребность перевести информацию, содержащуюся в таблице, к единому языку. В данном исследовании будет разумно перевести все на английский язык, так как данные с использованием кириллицы могут отображаться некорректно.

```python
# Средствами библиотеки traclator переводим
# информацию в таблице на английский язык
translator = Translator()
df['job_title']= df['job_title'].apply(lambda x: translator.translate(x, dest='en').text)
# Для применения перевода ко всему столбцу job_title
# используется простая лямбда-функция
```

```python
# Проверка
df['job_title'].value_counts()
```

```python
# Аналогично поступаем и со столбцом job_description,
# которые тоже является мультиязычным
df ['job_description']= df['job_description'].apply(lambda x: translator.translate(x, dest='en').text)
```

В данном шаге исследования вся информация в таблице была приведена к единому языку. Это упростит задачу по отсеиванию не-джунов и по вычленению hard skills.

## Исследовательский анализ данных

### Выборка из аналитиков согласно ТЗ

```python
# Смотрим список вакансий:
list(df['job_title'].unique())
```

Не все в списке вакансий являются аналитиками - одних можно отнести к вспомогательному персоналу, в других вакансиях ищут человека на позиции middle, senior. Согласно цели исследования нас интересуют только позиции джунов. Так как во многих вакансиях нет приставки Junior, будем считать, что они тоже являются джуновскими.

```python
# Сохраняем в список ключевые слова,
keywords = ['data analyst', 'science', 'bi-analyst', 
            'bi analyst', 'bi', 'business analyst', 
            'marketing analyst']
# по которым будем отбирать нужные нам вакансии
# средствами str.contains()
filtered_data = df[df['job_title'].str.contains('|'.join(keywords), case=False, na=False)]
```

```python
# Список слов, которые необходимо исключить
exclude_words = ['lead', 'team-lead', 
                 'team lead', 'senior', 
                 'middle', 'master']
# Создаем фильтр, который исключает строки, содержащие указанные слова
filter = ~filtered_data['job_title'].str.contains('|'.join(exclude_words), case=False)
# Применяем фильтр к датафрейму
filtered_data = filtered_data[filter]
```

```python
# Полученный результат:
filtered_data.info()
```

Таким образом, были отобраны джун-вакансии и результаты сохранены в переменную `filtered_data`.

### Выстаскиваем hard skills

```python
# Список хард скиллов
hard_skills = ["python", "database", "r", "sql", "data mining", "excel", "react", "statistic",
              'power bi', 'bi analysis', 'bi-analysis', 'tableau', 'oracle', 'etl',
              'mysql', 'postgresql', 'mongodb', 'pandas', 'numpy', 'scipy', 'matplotlib',
              'cloud', 'google metric', 'business intelligence', 'big data',
              'lean analytics', 'six sigma', 'machine learning', 'artificial intelligence', 'ai',
              'math', 'economy', 'marketing', 'gdpr', 'pr', 'public relations', 'cohort analysis',
              'cohort', 'powerpoint', 'data visualization', 'wrangling', 'data wrangling',
              'data manipulation', 'analytics', 'sas', 'spss', 'qlikview', 'nosql', 'presentation',
              'bpmn', 'aris', 'visio', 'jira', 'agile', 'kanban', 'trello', 'uml', 'erp', 'parsing',
              'eda', 'exploratory data analysis', 'seaborn', 'plotly', 'plotly express']

# Функция для вычленения хард скиллов из описания вакансии
def extract_hard_skills(description):
    extracted_skills = []
    for skill in hard_skills:
        if re.search(r'\b' + re.escape(skill) + r'\b', description, re.IGNORECASE):
            extracted_skills.append(skill)
    return extracted_skills

# Вычленение хард скиллов и сохранение их в отдельном столбце
filtered_data['hard_skills'] = filtered_data['job_description'].apply(lambda x: extract_hard_skills(x))

# Вывод таблицы
filtered_data.head()
```

Выборка именно таких hard skills обусловена всесторонним анализом европейского рынка труда и требований работодателей к умениями и навыкам на позицию джуниор. Список может расширятся по запросу Заказчика.

```python
# Не забываем обновлять индексы:
filtered_data = filtered_data.reset_index(drop=True)
filtered_data.head()
```
